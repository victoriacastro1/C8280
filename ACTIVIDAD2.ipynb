Escribe un script para dibujar esto en diferentes subplots.

```python
# Valores de x
x = np.linspace(start=-5, stop=5, num=150)

# Crear los subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Trazar las potencias en los subplots
axs[0, 0].plot(x, x, label='Lineal')
axs[0, 0].set_title('Potencia 1')
axs[0, 1].plot(x, x**2, label='Cuadrático')
axs[0, 1].set_title('Potencia 2')
axs[1, 0].plot(x, x**3, label='Cúbico')
axs[1, 0].set_title('Potencia 3')
axs[1, 1].plot(x, x**4, label='4ta potencia')
axs[1, 1].set_title('Potencia 4')

# Configurar leyendas y ajustes de diseño
for ax in axs.flat:
    ax.legend()
    ax.set_xlabel('x')
    ax.set_ylabel('y')

# Ajustar los espacios entre subplots
fig.tight_layout()

# Mostrar la figura con los subplots
plt.show()
```

 Escribe una script para dibujar a las primeras diez potencias de x con un título que diga x hasta N, donde N = 1,...,10 

```python
# Valores de x
x = np.linspace(-1, 1, 100)

# Crear la figura y los subplots
fig, axs = plt.subplots(5, 2, figsize=(10, 15))
fig.subplots_adjust(hspace=0.5)

# Trazar las potencias de x en los subplots
for n in range(10):
    row = n // 2  # Fila del subplot
    col = n % 2   # Columna del subplot
    power = n + 1  # Potencia correspondiente
    
    y = x**power
    axs[row, col].plot(x, y)
    axs[row, col].set_title(f'x hasta {power}')
    axs[row, col].set_xlabel('x')
    axs[row, col].set_ylabel(f'x^{power}')

# Mostrar la figura
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize = (14,4.5))
for i, ax in enumerate(axes.flatten()):
    ax.set_title("x hasta  {:d}".format(i+1))
    ax.plot(x, x**(i+1))
fig.tight_layout();
```

Supongamos que deseas que las etiquetas del eje x aparezcan en notaciones matemáticas como -pi y -pi/2. Escribe un script utilizando los métodos set_xticks, set_yticks, set_xticklabels y set_yticklabels para cambiar y asignar nuevos valores.

```python
# Generando datos
x = np.linspace(-np.pi, np.pi, 200)
seno, coseno = np.sin(x), np.cos(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()
ax.plot(x, seno, color='red')
ax.plot(x, coseno, color='#165181')

# Establecer límites de ejes
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-1.2, 1.2)

# Establecer ticks y etiquetas personalizadas para el eje x
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
xticklabels = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)

ax.set_yticks(np.arange(-1, 1.1, 0.5))

# Mostrar la figura
plt.show()
```

```python
# Generando datos
x = np.linspace(-np.pi, np.pi, 200)
seno, coseno = np.sin(x), np.cos(x)

# Crear la figura y los ejes
fig, ax = plt.subplots()
ax.plot(x, seno, color='red', label='Seno')
ax.plot(x, coseno, color='#165181', label='Coseno')

# Establecer límites de ejes
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-1.2, 1.2)

# Establecer ticks y etiquetas personalizadas para el eje x
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
xticklabels = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)

ax.set_yticks(np.arange(-1, 1.1, 0.5))

# Agregar etiquetas en la parte superior izquierda del gráfico
ax.text(-3, 0.8, 'Funciones', fontsize=12, fontweight='bold')
ax.text(-3, 0.6, 'Seno', color='red')
ax.text(-3, 0.4, 'Coseno', color='#165181')

# Mostrar la figura con leyenda
ax.legend()
plt.show()
```
